Da a la figura entera el título de "Mi restaurante"
</div>

```python
# funcion para sacar valores absolutos en grafico pie
tips_pie = tips.day

def autopct_fun(abs_values):
    gen = iter(abs_values)
    return lambda pct: f"{pct:.1f}% ({next(gen)})"

tips_pie_counts = tips_pie.value_counts()
```

```python
fig, ax = plt.subplots(2, 2)
fig.set_size_inches(8, 8)
plt.suptitle('Mi restaurante', fontsize=16)


#pie
ax[0, 0].pie(tips_pie_counts, autopct=autopct_fun(tips_pie_counts), labels = tips_pie_counts.index),
#histograma
sns.distplot(tips["tip"], bins = 20, kde = False, ax = ax[0, 1]),
#violin
sns.violinplot(data = tips, x = "day", y = "size", ax = ax[1, 0]);
# dispersion + regresion  lmplot regplot
sns.regplot(x="tip", y="total_bill", data=tips, ax = ax[1, 1]),
ax[1, 1].set_xticks(range(0, 11, 1))
plt.show();


```

Partiendo de la versión "reducida" de nuestro dataset (trips_reduced), queremos ahora analizar la evolución del número de viajes a lo largo de los diferentes meses (de los dos años para los que tenemos datos). Para ello deberemos agrupar nuestros datos por meses, lo que haremos no usando el método .groupby, sino usando el método .resample (orientado al uso con series temporales). Para la agrupación, considera el mes del campo "Start date" como período de agrupación, y calcula el recuento de alguno de los campos (por ejemplo del campo "Duration"). Guarda el resultado en la variable "time_evolution" y muestra en una gráfica la evolución obtenida.
</div>

```python
time_evolution = trips_reduced.set_index('Start date')
time_evolution
```

```python
time_evolution = time_evolution.resample('M').count()
```

```python
time_evolution
```

```python
#representación con pandas
time_evolution["Duration"].plot.line();
```

```python
#representación con matplotlib
fig, ax = plt.subplots()
plt.plot(time_evolution[["Duration"]])
plt.title("Time Evolution")
plt.xlabel("Months")
plt.xticks(rotation=45)
plt.ylabel("Frecuency")
plt.show();
```
