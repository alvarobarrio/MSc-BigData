Por ltimo, en el conjunto de ejes situado a la derecha, crea un grfico de dispersin cruzando las variables "tip" y "total_bill" que incluya una recta de regresin para los datos mostrados. Asegrate de que se muestran "ticks" para todos los valores entre 0 y 10, ambos incluidos.
    
Da a la figura entera el ttulo de "Mi restaurante"
</div>

```python
# funcion para sacar valores absolutos en grafico pie
tips_pie = tips.day

def autopct_fun(abs_values):
    gen = iter(abs_values)
    return lambda pct: f"{pct:.1f}% ({next(gen)})"

tips_pie_counts = tips_pie.value_counts()
```

```python
fig, ax = plt.subplots(2, 2)
fig.set_size_inches(8, 8)
plt.suptitle('Mi restaurante', fontsize=16)


#pie
ax[0, 0].pie(tips_pie_counts, autopct=autopct_fun(tips_pie_counts), labels = tips_pie_counts.index),
#histograma
sns.distplot(tips["tip"], bins = 20, kde = False, ax = ax[0, 1]),
#violin
sns.violinplot(data = tips, x = "day", y = "size", ax = ax[1, 0]);
# dispersion + regresion  lmplot regplot
sns.regplot(x="tip", y="total_bill", data=tips, ax = ax[1, 1]),
ax[1, 1].set_xticks(range(0, 11, 1))
plt.show();


```

Partiendo de la versin "reducida" de nuestro dataset (trips_reduced), queremos ahora analizar la evolucin del nmero de viajes a lo largo de los diferentes meses (de los dos aos para los que tenemos datos). Para ello deberemos agrupar nuestros datos por meses, lo que haremos no usando el mtodo .groupby, sino usando el mtodo .resample (orientado al uso con series temporales). Para la agrupacin, considera el mes del campo "Start date" como perodo de agrupacin, y calcula el recuento de alguno de los campos (por ejemplo del campo "Duration"). Guarda el resultado en la variable "time_evolution" y muestra en una grfica la evolucin obtenida.

```python
time_evolution = trips_reduced.set_index('Start date')
time_evolution
```

```python
time_evolution = time_evolution.resample('M').count()
```
